### *Data Visualization*

Visualization of the Variables 

```python
sns.countplot(x='Ship Mode', data=s_df)
plt.show()
```

```python
sns.countplot(x='Segment', data=s_df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x=s_df['State'])
plt.xticks(rotation=90)
plt.title('Count of State wise shipments')
plt.show()
```

State wise California has highest sales. 

States like Texas, New York, etc are active in business. 

Few states like North Dakota, Wyoming and West Virginia have less or no sales at all. 

To generate more profit we have to increase our sales in Texas and New York where we have potential customers.

```python
plt.figure(figsize = (6,6))
textprops = {"fontsize":15}
plt.title('Category')
plt.pie(s_df['Category'].value_counts(), labels=s_df['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

This proves Office supplies have highest demand and others are comparatively less.

```python
s_df.groupby('Region')['Profit'].agg(sum).plot(kind='bar')
plt.ylabel('Profit')
plt.title('Profits in different regions')
plt.show()
```

West generates most of the profit followed by East. 

Central and Southern region are generating less profit as compared to the Eastern and Western Region. 

Increase the sales in Central and Southern region in order to generate more profit in these region.

```python
sp = s_df.groupby('State')['Profit'].sum()
plt.figure(figsize=(16,6))
sns.barplot(x=sp.index, y=sp.values)
plt.xticks(rotation=90)
plt.ylabel('Profit')
plt.show()
```

```python
sns.pairplot(s_df, hue='Segment')
```

### *Heat map showing the correlation between the features*

```python
sns.heatmap(s_df.corr(), annot = True, cmap = "PiYG")
```

```python
s_df.groupby(['Category'])[['Profit','Sales']].agg(sum).plot(kind='bar',figsize=(10,4), )
plt.ylabel('Profit and Sales')
plt.show()
```

This proves that Furniture is generating very less profit.

```python
plt.figure(figsize=(9,5))
sns.countplot(x='Category',hue='Segment',data=s_df)
plt.ylabel('Count of Segment')
plt.title('Count of Segment by Category and Segment')
```

The highest purchases are in consumer segment whereas home office segment has the least purchases in each category. We need to focus on the corporate and home office segment as well to increase our sales and profit.

```python
s_df.groupby('Sub-Category')[['Sales','Profit']].agg(sum).plot(kind='barh',figsize=(15, 8))
plt.ylabel('Sub-Category')
plt.xlabel('Sales and Profit')
plt.show()
```

We have 3 weak areas Bookcases, Supplies and Tables as we are facing loss on the sales of these items. On the sales of tables we are facing highest loss.

```python
plt.figure(figsize=(9,5))
sns.countplot(x='Region',hue='Ship Mode',data=s_df)
plt.ylabel('Count of Ship Mode')
plt.title('Count of ship mode by region and ship mode')
```

Least sales are in Southern region, we can attract this region by offering more discount. Standard mode of shipment is opted by most of the customers in each region.

```python
s_df.plot(kind = "scatter",figsize = (10,5), x = "Sales", y= "Profit", c = "Discount", s = 20,fontsize = 16, colormap = "viridis")
plt.ylabel('Total Profits',fontsize = 16)
plt.title("Interdependency of Sales,Profits and Discounts",fontsize = 16)
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.lineplot(s_df['Discount'], s_df['Profit'], data=s_df)
```

To keep the profit margins the percentage of discount should be less than 25%.
